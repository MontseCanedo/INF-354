Se importan las librerías y se cargan los datos

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
cc_apps = pd.read_csv("cca.csv", header=None)
```

Se imprime la información del dataset. Se hace la revision de valores faltantes el en dataset y se reemplaza por NaN. Se verifica de nuevo los datos faltantes

```python
x = cc_apps
y = [0,1]
print(cc_apps.head())
cc_apps
print("\n")
cc_apps_info = cc_apps.info()
print(cc_apps_info)
print("\n")
cc_apps.tail(17)
print(cc_apps.isnull().values.sum())
cc_apps = cc_apps.replace("?",np.NaN)
cc_apps.tail(17)
```

Muestra la verificación de las iteraciones por columna contando los valores Nan

```python
cc_apps = cc_apps.fillna(cc_apps.mean())
print(cc_apps.isnull().values.sum())
print(cc_apps.info())

for col in cc_apps.columns:
    if cc_apps[col].dtypes == 'object':
        """Verifica si la columna es de tipo object"""
        cc_apps[col] = cc_apps[col].fillna(cc_apps[col].value_counts().index[0])
        """lo representa con el valor más frecuente"""
print(cc_apps.isnull().values.sum())
```

Se itera en todos los valores de cada columna y se extrae sus dtypes. Se convierte el dataset en una matriz Numpy y separa etiquetas en variables separadas

```python
le = LabelEncoder()
for col in cc_apps.columns:
    if cc_apps[col].dtype=='object':
        cc_apps[col]=le.fit_transform(cc_apps[col])
cc_apps = cc_apps.drop([cc_apps.columns[10],cc_apps.columns[13]], axis=1)
cc_apps
cc_apps = cc_apps.values
X,y = cc_apps[:,0:13], cc_apps[:,13]
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
rescaledX = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(rescaledX,
                                                    y,
                                                    test_size=0.20,train_size=0.80,
                                                    random_state=42)
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_test)
print("Precisión del clasificador de regresión logística: ", logreg.score(X_test, y_test))
print(confusion_matrix(y_test, y_pred))
print("\n Ejemplos usados para entrenar TRAIN: ")
print("Ejemplos usados para test: TEST\n")
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print('\n X_train: ',len(X_train))
print(X_train)
print('\n X_test: ',len(X_test))
print(X_test)
print('y_train: ',len(y_train))
print(y_train)
print('\n y_test: ',len(y_test))
print(y_test)

classifier = MLPClassifier(activation="relu",solver='adam', hidden_layer_sizes=(100,),max_iter=300)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
print('\n Y_PREDICC: ',y_pred)
accuracy = accuracy_score(y_test, y_pred)
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Datos estadísticos de las precisiones\n")
print('Media de las precisiones después de la validación cruzada: ', accuracies.mean())
print('Desviación estándar dentro de las precisiones: ', accuracies.std())
print('precisiones: ', accuracies)
```
